## **Question 1: What is Dimensionality Reduction? Why is it important in machine learning?**

Dimensionality Reduction is the process of reducing the number of features or variables in a dataset while preserving essential information. In many real-world datasets, especially those involving images, text, or sensor data, the number of features can be extremely high. This high dimensionality increases computational cost, causes model overfitting, and makes visualization difficult. This phenomenon is known as the curse of dimensionality.  
Dimensionality reduction simplifies data, removes noise, and improves machine learning performance by focusing only on meaningful patterns. Techniques like PCA, t-SNE, and LDA help convert high‑dimensional data into a compact, informative representation. This makes models faster, more accurate, and easier to interpret.

## **Question 2: Name and briefly describe three common dimensionality reduction techniques.**

1. **PCA (Principal Component Analysis):**  
   Converts high‑dimensional data into a smaller set of principal components that capture maximum variance.

2. **t-SNE (t‑Distributed Stochastic Neighbor Embedding):**  
   A nonlinear technique that visualizes high-dimensional data in 2D or 3D while preserving local similarity.

3. **LDA (Linear Discriminant Analysis):**  
   A supervised method that reduces dimensions by maximizing class separability.

## **Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.**

Clustering is an unsupervised learning technique that groups similar data points together based on shared patterns without using labeled outputs. The goal is to minimize intra‑cluster similarity and maximize inter‑cluster separation. Clustering is widely used in customer segmentation, anomaly detection, image grouping, and pattern identification.

Three popular clustering algorithms:  
- **K-Means Clustering** – Partitions data into K clusters using centroid optimization.  
- **Hierarchical Clustering** – Builds a tree (dendrogram) of merging/splitting clusters.  
- **DBSCAN** – Density-based method that detects clusters of arbitrary shapes and identifies noise/outliers.

## **Question 4: Explain the concept of anomaly detection and its significance.**

Anomaly detection identifies rare, unusual, or suspicious patterns in data that differ from normal behavior. These anomalies could indicate fraud, system faults, cyber-attacks, equipment failure, or medical abnormalities.  
It is significant because anomalies often represent critical events that require immediate attention. In industries like finance, healthcare, manufacturing, and cybersecurity, early detection helps prevent losses, improve safety, and maintain reliability. Techniques such as statistical models, machine learning, and deep learning help automate anomaly detection across large datasets.

---

## **Question 5: List and briefly describe three types of anomaly detection techniques.**

1. **Statistical Methods:**  
   Assume normal data follows a known distribution. Outliers far from the mean or expected probability range are flagged.

2. **Machine Learning Methods:**  
   Algorithms like Isolation Forest, One-Class SVM, and LOF learn normal patterns and detect deviations.

3. **Deep Learning Methods:**  
   Autoencoders and LSTM networks learn complex patterns and detect anomalies using reconstruction errors or prediction deviations.

## **Question 6: What is time series analysis? Mention two key components of time series data.**

Time series analysis involves examining data collected over time to identify patterns, trends, and seasonal fluctuations. It is used in forecasting, anomaly detection, and decision-making across finance, weather, manufacturing, and health.

Two key components:  
- **Trend:** Long-term upward or downward movement in data.  
- **Seasonality:** Repeating patterns at regular intervals (daily, monthly, yearly).

---

## **Question 7: Describe the difference between seasonality and cyclic behavior in time series.**

**Seasonality** refers to regular, predictable patterns that repeat at fixed intervals, such as daily sales cycles or yearly weather effects.  
**Cyclic behavior**, however, occurs over irregular, longer periods and is influenced by economic, social, or natural factors. Unlike seasonality, cycles do not have fixed durations and are harder to predict.

## **Question 8: Python code to perform K-means clustering on a sample dataset**


```python

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate dataset
X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# K-Means
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
labels = kmeans.predict(X)

# Plot
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='X', s=200)
plt.title("K-Means Clustering Output")
plt.show()
```
