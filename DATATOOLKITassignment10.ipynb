1. What is NumPy, and why is it widely used in Python?

2. How does broadcasting work in NumPy?

Broadcasting in NumPy allows arithmetic operations between arrays of different shapes. Instead of explicitly replicating smaller arrays, NumPy stretches them virtually to match the larger array shape. For example, adding a scalar to a 2D array applies the operation element-wise without looping. Broadcasting follows rules: dimensions are compared from right to left, and if they differ, one dimension is expanded if it is 1. This makes calculations concise, memory-efficient, and fast. It eliminates writing complex loops, which improves readability. Broadcasting is fundamental in vectorized operations, making NumPy powerful for scientific and data-intensive computations.

6. What are the differences between NumPy arrays and Python lists?

NumPy arrays and Python lists differ in structure, efficiency, and functionality. Lists are more flexible for general-purpose storage, but arrays are better for numerical and scientific computing. 8. What does the term “vectorized operation” mean in NumPy?

A vectorized operation in NumPy means performing operations on entire arrays without writing explicit Python loops. Instead of iterating over elements one by one, NumPy applies mathematical operations across all elements simultaneously using optimized C implementations. For example, adding two arrays directly performs element-wise addition much faster than looping through each index. Vectorization makes code shorter, easier to read, and significantly faster, especially for large datasets. It also supports broadcasting, allowing operations between arrays of different shapes. This efficiency is one of the key reasons NumPy is preferred for numerical and scientific computations.

17. Explain the difference between apply() and map() in Pandas.

In Pandas, map() is used on Series to apply a function element-wise, often for simple transformations or replacements. For example, replacing string values or applying a mathematical operation to each element. The apply() method works on both DataFrames and Series, allowing more complex operations along rows or columns. apply() can handle custom functions and even lambda expressions for advanced transformations. While map() is limited to Series, apply() is more versatile for DataFrames. In short, map() is suitable for simple, one-dimensional operations, whereas apply() is more powerful and flexible for multi-dimensional transformations.

21. Why is NumPy’s array slicing faster than Python’s list slicing?

NumPy’s array slicing is faster because arrays are stored in contiguous blocks of memory with a fixed data type. This allows efficient indexing and operations at the C level without Python’s overhead. In contrast, Python lists store references to objects, which introduces additional memory lookups. NumPy uses strides to create “views” of slices without copying data, further improving performance. This design enables vectorized operations and reduces computation time. The efficiency of array slicing is one of the reasons why NumPy is preferred for numerical tasks over Python’s built-in list structures.

```python
# Q13: Create a 3D Scatter Plot
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data
x = [1, 2, 3, 4, 5]
y = [5, 3, 6, 2, 7]
z = [2, 3, 5, 7, 1]

# Create 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c='red', marker='o')

# Labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
ax.set_title('3D Scatter Plot')

# Display plot
plt.show()
```
