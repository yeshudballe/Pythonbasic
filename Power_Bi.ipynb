Question 1 :
Define Power BI and What are the key components of the Power BI ecosystem? Briefly
-
explain:
* Power BI Desktop
* Power BI Service
* Power BI Mobile
* Power BI Gateway

Answer:-

Power BI is a business analytics tool developed by Microsoft that helps users collect, analyze, visualize, and share data through interactive dashboards and reports. It supports data-driven decision making by converting raw data into meaningful insights.
1) Power BI Desktop

A free Windows application used to create reports.

Used for:

 - Connecting to data sources

 - Data cleaning & transformation (Power Query)

 - Creating charts and visuals

 - Reports created here are published to Power BI Service.

2) Power BI Service

* A cloud-based online platform.
* Used for:
  - Sharing reports and dashboards

  - Collaboration with team members

  - Scheduling data refresh
* Accessible through web browser.

3) Power BI Mobile

* Mobile app available for Android & iOS.
* Used to:
  - View dashboards & reports on phone

  - Get real-time alerts

  - Access data anywhere.

4) Power BI Gateway
* Acts as a bridge between on-premise data and cloud.
* Used when:
  - Data is stored locally (SQL Server, Excel, etc.)
  - Secure data transfer to Power BI Service
* Types:
  - Personal Mode
  - Enterprise Mode


Question 2 : Compare the following Power BI visuals:
-

*  Pie Chart vs Donut Chart
*  Bar Chart vs Column Chart
When would you prefer one over the other? Give one example for each pair.
Answer:

| Feature      | Pie Chart                        | Donut Chart                             |
| ------------ | -------------------------------- | --------------------------------------- |
| Shape        | Solid circle                     | Circle with center hole                 |
| Data Display | Shows part-to-whole relationship | Same as pie but with better readability |
| Space        | Uses full area                   | Center space can show total value       |
| Look         | Simple                           | More modern & attractive                |

 When to Use
* Pie Chart:

Use when you have 2–5 categories and want to show simple percentage distribution.

* Donut Chart:

Use when you want to display total value in center and better design.

Example

 - Pie Chart: Market share of mobile brands (Samsung, Apple, Xiaomi).

 - Donut Chart: Monthly expense distribution with total amount shown in center.

| Feature     | Bar Chart                        | Column Chart                |
| ----------- | -------------------------------- | --------------------------- |
| Direction   | Horizontal                       | Vertical                    |
| Best for    | Long category names              | Time-based data             |
| Readability | Easy to read long labels         | Better for trend comparison |
| Usage       | Compare values across categories | Show changes over time      |

When to Use

* Bar Chart:

Use when category names are long or many items.

* Column Chart:

Use when you want to show trend or growth over time.

Example

 - Bar Chart: Comparison of marks in different subjects.

 - Column Chart: Monthly sales report of a company.

Question 3: Explain the significance of:
-
- Star schema vs Snowflake schema
- Primary key vs Foreign key in relationships (Power BI)

Why is cardinality important?

Answer:-

1) Star Schema vs Snowflake Schema

* Star Schema

  - Structure:
One central fact table connected to multiple dimension tables.

  - Shape:
Looks like a star.

  - Advantages:

Simple design

Faster query performance

Easy to understand

  - Usage:

Preferred in Power BI and data warehousing.

  - Example:
Sales Fact → Customer, Product, Date, Store



* Snowflake Schema

 - Structure:
Fact table → Dimension tables → Sub-dimensions

 - Shape:
Looks like a snowflake.

 - Advantages:

Reduces data redundancy

Saves storage

  - Disadvantages:

Complex joins

Slower performance

  - Usage:

Used when data is highly normalized.

| Feature     | Star Schema | Snowflake Schema |
| ----------- | ----------- | ---------------- |
| Complexity  | Simple      | Complex          |
| Performance | Fast        | Slower           |
| Storage     | More        | Less             |
| Maintenance | Easy        | Hard             |

2) Primary Key vs Foreign Key (Power BI Relationships)

- Primary Key

   - A unique identifier in a table.

   - Cannot contain duplicate or NULL values.

   - Used to uniquely identify records.

Example:
CustomerID in Customer table

3) Why is Cardinality Important?

- Cardinality defines the relationship type between tables:

 - One-to-One (1:1)

  - One-to-Many (1:*)

  - Many-to-Many (:)

- Importance

  - Ensures correct data matching

  - Affects filtering and aggregation

  -  Improves report accuracy

  - Prevents duplicate or incorrect totals

Example:
One Customer → Many Orders (1:*)

Question 4 :Differentiate between:
-
- Calculated column vs Measure
- Also, define Row context and Filter context with simple examples.

Answer:-

1) Calculated Column vs Measure

| Feature          | Calculated Column                  | Measure                            |
| ---------------- | ---------------------------------- | ---------------------------------- |
| Calculation time | Calculated **when data is loaded** | Calculated **at run-time**         |
| Stored in table  | Yes (takes memory)                 | No (calculated dynamically)        |
| Use case         | Row-level calculations             | Aggregations & KPIs                |
| Used in          | Table visuals, filters             | Cards, charts, dashboards          |
| DAX example      | `Profit = Sales - Cost`            | `Total Sales = SUM(Sales[Amount])` |


2) Row Context

Definition:
Row context means calculation is done row by row.

Example

Sales Table:

| Price | Quantity |
| ----- | -------- |
| 100   | 2        |
| 200   | 3        |


Calculated Column:

     Total = Price * Quantity


Power BI calculates each row separately:

100 × 2 = 200

200 × 3 = 600


3) Filter Context

Definition:
Filter context means calculation changes based on filters/slicers.

Example

Measure:

     Total Sales = SUM(Sales[Amount])


Without filter → Shows total of all sales

With filter (Year = 2025) → Shows only 2025 sales

Question 5:What is the difference between a report and a dashboard in Power BI?
-
Answer:

Power BI Report

- Definition:
A report is a collection of visuals created from one or more datasets.

- Key Points:

  - Created in Power BI Desktop

  - Can have multiple pages

  - Interactive (filters, slicers)

  - Used for detailed analysis

  - Editable by creator

- Example:
Sales performance report with pages:

  - Monthly sales

  - Product-wise sales

  - Region-wise sales

Power BI Dashboard

- Definition:
A dashboard is a single-page summary view created in Power BI Service.

- Key Points:

  - Contains tiles pinned from different reports

  - Shows key KPIs

  - Read-only for users

  - Used for quick monitoring

  - Can combine multiple datasets

- Example:
  - Business overview dashboard:

  - Total revenue

  - Top-selling product

  - Monthly growth
