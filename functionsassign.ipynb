1. What is the difference between a function and a method in Python?
  - In Python, both functions and methods are used to perform specific operations, but they differ in how they are used. A function is a block of reusable code that performs a specific task and is defined independently, usually using the def keyword. It can be called anywhere in the program with appropriate arguments. A method, on the other hand, is associated with an object and is called using that object. Methods are essentially functions that belong to objects of classes. They automatically receive the object as their first argument, which is commonly referred to as self. While functions can stand alone, methods are tied to the object’s type and can access and modify its data. In short, every method is a function, but not every function is a method. This distinction becomes important especially when working with object-oriented programming concepts in Python, where methods define object behaviors.

2. Explain the concept of function arguments and parameters in Python.
  - In Python, function parameters and arguments are both essential for passing data into functions, but they represent different aspects. Parameters are the variable names listed in the function’s definition, acting as placeholders for the data that will be passed. Arguments, on the other hand, are the actual values provided when the function is called. When a function is executed, each parameter is assigned the value of its corresponding argument. Python also allows different types of arguments such as positional, keyword, default, and variable-length arguments. Positional arguments match parameters by their order, while keyword arguments explicitly mention the parameter names. Default arguments provide a fallback value when no argument is passed, and variable-length arguments allow passing multiple values. Understanding the relationship between parameters and arguments is crucial for writing flexible and reusable functions that can handle various types of input and perform different operations accordingly.

3. What are the different ways to define and call a function in Python?
  - In Python, functions can be defined using the def keyword followed by the function name and parameters inside parentheses. Functions can also be created using lambda expressions for short, anonymous functions. Functions are called by using their name followed by parentheses, optionally including any required arguments. There are multiple ways to call a function: directly by name, through other functions, inside loops, or by using built-in functions like map, filter, and reduce. Python also supports higher-order functions, which means functions can accept other functions as arguments or return functions as results. Additionally, recursive functions call themselves during execution. The flexibility in defining and calling functions allows Python programmers to create highly modular, reusable, and readable code. Whether simple or complex, functions help organize logic into manageable sections, and understanding their usage is vital to effective Python programming.

4. What is the purpose of the return statement in a Python function?
  - The return statement in a Python function is used to send a value back to the caller once the function has completed its task. It terminates the function's execution and provides the result to the place where the function was invoked. If no return statement is present, the function returns None by default. The return statement is essential for functions that need to output a result rather than just perform an action. It allows data to be passed from inside the function to the outside, making it usable in other parts of the program. A function can have multiple return statements, typically used in conditional branches, but it only executes the first one it encounters during a call. Using return values effectively makes a function reusable and helps in building more modular code structures. It is one of the core aspects of designing meaningful and productive functions in Python.

5. What are iterators in Python and how do they differ from iterables?
  - In Python, iterables and iterators are key components of loops and data traversal. An iterable is any object capable of returning its elements one at a time, allowing it to be looped over using a for loop. Examples include lists, strings, tuples, and dictionaries. However, these objects do not keep track of their current position during iteration. An iterator, in contrast, is an object that implements the iterator protocol, which includes the methods __iter__() and __next__(). Once an iterable is converted into an iterator, it remembers the position and returns the next value each time it is called until all items are exhausted. The main difference is that iterators can be manually controlled and advanced, whereas iterables are just collections of elements. Iterators are especially useful for working with streams of data or infinite sequences where keeping track of the current state is necessary.

6. Explain the concept of generators in Python and how they are defined.
  - Generators in Python are a type of iterable, like lists or tuples, but unlike those, they do not store all values in memory at once. Instead, they generate items one by one using a special type of function. Generators are defined using regular function syntax but with the yield keyword instead of return. When a generator function is called, it returns a generator object without executing the function's code immediately. Each time the generator is iterated, it runs the function until it reaches a yield, pauses, and returns the yielded value. When resumed, it continues from where it left off. This behavior makes generators highly memory-efficient, especially when dealing with large datasets or infinite sequences. Generators are commonly used in situations where we want to compute values on demand rather than all at once, and they allow us to write cleaner and more readable code for such tasks.

7. What are the advantages of using generators over regular functions?
  - Generators offer several advantages over regular functions, particularly in terms of memory usage and efficiency. Unlike regular functions that return all results at once and store them in memory (e.g., as a list), generators produce values one at a time as needed using the yield keyword. This makes them more memory-efficient, especially useful when working with large datasets or streams of data. Generators also allow for lazy evaluation, meaning they compute values only when required, leading to performance benefits. They can be paused and resumed, making them suitable for applications that need to maintain state between operations. Additionally, generators simplify code when dealing with pipelines or sequences that require multiple steps of processing. By using generators, programmers can build more responsive and scalable programs that handle data more effectively without overloading system resources, which would be difficult to achieve using regular functions that generate all results at once.

8. What is a lambda function in Python and when is it typically used?
  - A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions created with def, lambda functions are meant for short, simple operations and are typically written in a single line. They do not have a name (unless assigned to a variable) and are used where a quick, throwaway function is needed. Lambda functions are most commonly used in situations that require a function as an argument, such as with map(), filter(), or sorted(). They are useful for simplifying code that would otherwise require a full function definition, making the logic more concise and readable. Although limited in complexity, lambda functions improve efficiency in cases where defining a full function is unnecessary. However, they are not a replacement for more complex or reusable functions and are best used for simple expressions where readability is maintained.

9. Explain the purpose and usage of the map() function in Python.
  - The map() function in Python is used to apply a given function to all the items in an iterable, such as a list or tuple, and return a new iterable (specifically, a map object). Its primary purpose is to simplify and streamline operations that need to be applied to multiple elements without using explicit loops. map() takes two arguments: the function to apply and the iterable to process. The function is executed on each element, and the result is collected. This is especially useful for transforming data, such as converting numbers, changing string formats, or applying calculations. The output can be easily converted into a list, tuple, or any other iterable type. Using map() helps improve code readability and efficiency, particularly in cases where the same operation must be performed on every element of a collection. It is often used with lambda functions for concise, one-line transformations.

10. What is the difference between map(), reduce(), and filter() functions in Python?
  - map(), reduce(), and filter() are built-in Python functions used for functional programming, each serving a different purpose when working with iterables. map() applies a function to every item in an iterable and returns a new iterable with the transformed items. It is used when we want to modify or process all elements. filter() also applies a function, but it only keeps the elements for which the function returns true, effectively filtering the original data. reduce() is used to apply a function cumulatively to the items of a sequence, reducing them to a single value, such as summing or multiplying all elements. Unlike map() and filter(), reduce() must be imported from the functools module. While map() and filter() return iterables, reduce() returns a single value. All three functions support concise and readable code when dealing with repetitive data transformation, aggregation, or filtering operations in Python.

```python
# PRACTICAL QUESTIONS.
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

print(sum_even_numbers([1, 2, 3, 4, 5, 6]))
```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.
def reverse_string(text):
    reversed_text = ""
    for char in text:
        reversed_text = char + reversed_text
    return reversed_text

print(reverse_string("hello"))
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_list(numbers):
    squared = []
    for num in numbers:
        squared.append(num * num)
    return squared

print(square_list([1, 2, 3, 4]))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

for i in range(1, 201):
    if is_prime(i):
        print(i, "is prime")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, limit):
        self.a = 0
        self.b = 1
        self.count = 0
        self.limit = limit

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.limit:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

for num in Fibonacci(10):
    print(num)
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for value in powers_of_two(5):
    print(value)

```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_lines(filename):
    try:
        file = open(filename, 'r')
        for line in file:
            yield line.strip()
        file.close()
    except FileNotFoundError:
        print("File not found!")

# Use the function like this:
for line in read_lines("sample.txt"):
    print(line)

```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(1, 3), (4, 1), (2, 5)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 10, 20, 30]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(text):
    vowels = "aeiouAEIOU"
    result = filter(lambda ch: ch not in vowels, text)
    return "".join(result)

print(remove_vowels("Hello World"))
```
