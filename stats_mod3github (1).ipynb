### Question 1: What is hypothesis testing in statistics?


**Answer:**  
Hypothesis testing is a statistical procedure used to make inferences about populations based on sample data. It involves formulating two competing statements: the null hypothesis (H₀), which assumes no effect or difference, and the alternative hypothesis (H₁), which proposes a significant effect or difference. By analyzing sample data, statistical tests calculate probabilities (p-values) to decide whether to reject H₀ in favor of H₁. Hypothesis testing provides a structured framework to reduce bias, manage uncertainty, and guide data-driven decisions in research, quality control, and other fields.


### Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?


**Answer:**  
The null hypothesis (H₀) is a default assumption stating there is no effect, relationship, or difference in the population. For example, H₀ might claim a medicine has no impact on blood pressure. The alternative hypothesis (H₁ or Ha) contradicts H₀ and suggests a real effect or difference exists, such as the medicine reducing blood pressure. Hypothesis testing evaluates whether sample data provides enough evidence to reject H₀. While H₀ is conservative and assumes status quo, H₁ represents the researcher’s claim, tested statistically.


### Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.


**Answer:**  
The significance level, denoted as α, is the threshold probability used in hypothesis testing to decide whether to reject the null hypothesis. Common values include 0.05, 0.01, and 0.10. It represents the maximum acceptable risk of committing a Type I error, i.e., rejecting H₀ when it is true. If the calculated p-value is less than α, H₀ is rejected in favor of H₁, suggesting strong evidence against the null. A smaller α reduces false positives but increases the chance of Type II errors. Choosing α balances risk and consequences.


### Question 4: What are Type I and Type II errors? Give examples of each.


**Answer:**  
Type I error occurs when the null hypothesis (H₀) is rejected despite being true. For example, concluding a new drug is effective when it actually is not. The probability of Type I error is α (significance level). Type II error occurs when H₀ is not rejected even though it is false. For example, failing to detect the effectiveness of a real treatment. Its probability is β, and power = 1 – β represents the chance of correctly rejecting H₀. Both errors highlight the trade-off between avoiding false positives and false negatives.


### Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.


**Answer:**  
A Z-test is used when the population standard deviation is known, or when the sample size is large (n > 30), assuming normal distribution. It is commonly applied in large-sample mean or proportion comparisons. A T-test, on the other hand, is used when the population standard deviation is unknown, and the sample size is small (n ≤ 30). It relies on the Student’s t-distribution, which accounts for extra uncertainty in smaller samples. Both tests assess whether observed sample means differ significantly from expected values, with T-tests being more flexible for small datasets.


### Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python

import numpy as np
import matplotlib.pyplot as plt

# Parameters
n, p, size = 10, 0.5, 1000

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor='black', density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

### Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python.

```python

import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mean = np.mean(sample_data)
std = np.std(sample_data, ddof=1)
n = len(sample_data)
mu0 = 50  # Hypothesized mean

z = (mean - mu0) / (std/np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print("Sample Mean:", mean)
print("Z-statistic:", z)
print("p-value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis: Mean differs significantly from 50")
else:
    print("Fail to Reject Null Hypothesis: No significant difference from 50")
```

### Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate normal distribution
data = np.random.normal(loc=50, scale=5, size=1000)

# 95% Confidence Interval
mean = np.mean(data)
sem = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print("Mean:", mean)
print("95% CI:", ci)

# Plot data
plt.hist(data, bins=30, edgecolor='black', density=True)
plt.axvline(ci[0], color='red', linestyle='dashed', linewidth=2)
plt.axvline(ci[1], color='red', linestyle='dashed', linewidth=2)
plt.title("Normal Distribution with 95% CI")
plt.show()
```

### Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram.

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generate random dataset
data = np.random.normal(100, 15, 500)

# Calculate Z-scores
z_scores = zscore(data)

# Plot histogram of standardized data
plt.hist(z_scores, bins=30, edgecolor='black', density=True)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.show()

print("First 10 Z-scores:", z_scores[:10])
print("Interpretation: A Z-score shows how many standard deviations a value is from the mean.")
```
